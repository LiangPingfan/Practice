```python
import numpy as np
import pandas as pd

df = pd.read_csv('X_train',header = 0,dtype = int)





x = df.iloc[:,1:]

x
```

```python
a = pd.read_csv('Y_train',header = 0)
y = a.iloc[:,1:]

y.head()
```

```python
np.zeros([3,2])
```

```python
mean_x = np.mean(x, axis = 0)
std_x = np.std(x,axis = 0)
for i in range(len(x['age'])):
    x['age'][i] = (x['age'][i] - mean_x['age'])/std_x['age']
```

```python
dim = x.shape[1]  
w = np.zeros([dim,1])

rate = 100
iter_time = 1000
adagrad = np.zeros([dim,1])
eps = 0.00000000000001
for i in range(iter_time):
    z = np.dot(x,w)
    fx = 1/(1+np.e**(-z))
    
    c = np.sum(- (np.dot(y,np.log(fx).transpose()))
               +np.dot((1-y),np.log(1-fx).transpose()))
    gradient = -np.dot((y-fx),x)
    adagrad += gradient ** 2
    w = w - rate*gradient/np.sqrt(adagrad + eps)
np.save('weight.npy',w)
```
